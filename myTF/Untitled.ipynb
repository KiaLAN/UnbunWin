```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

times = 100
colors = [cm.coolwarm(x) for x in np.linspace(-1,1,times)]
for i in range(times):
    plt.plot(np.random.rand(2)+i, np.random.rand(2),c=colors[i])

plt.show()
```
